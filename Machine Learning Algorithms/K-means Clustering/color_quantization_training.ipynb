```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
palmimg = plt.imread('./palm_trees.jpg')
```

```python
plt.imshow(palmimg)
```

```python
w, h, c = palmimg.shape
```

```python
palmimg = palmimg.reshape((w*h, c))
```

```python
palmimg.shape
```

```python
from sklearn.cluster import KMeans

model = KMeans(20)
model.fit(palmimg)
```

```python
labels = model.labels_
```

```python
colors = model.cluster_centers_.round().astype(int)
```

```python
labels
```

```python
finalimg = np.reshape(colors[labels], (w, h, c))
```

```python
plt.imshow(finalimg)
```
